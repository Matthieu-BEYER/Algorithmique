<div class="alert alert-warning">   <div style="text-align:justify;"> Lorsque l'on doit répéter la même action un certain nombre de fois ou bien jusqu'à ce qu'un évènement se produise on utilise une <FONT color="red"> boucle </FONT>. Python offre la possibilité d'utiliser deux types de boucle :

 ### <p style="background-color:silver;"> <b> <u> a) Utilisation de la boucle for </u> </b> </p>

<div class="alert alert-success" >  <div style="text-align:justify;"> Propriété 1 :  <br>
 La syntaxe à utiliser pour répéter un bloc d'instruction un nombre n de fois connu l'avance : <br>
 <code>for i in range(n):
        instructions </code>
    
</div>

<div class="alert alert-warning">   <div style="text-align:justify;"> Remarques :
  <ul>
<li> les deux points en fin de première ligne introduisent le bloc d'instructions à répéter </li>
<li> les instructions sont indentés (décaler vers la droite en appuyant sur la touche tabulation) </li>
<li> la fin de l'indentation signale la fin du bloc d'instruction à répéter. On sort de la boucle. </li>
<li> on peut imbriquer plusieurs boucle for </li>     

<div class="alert alert-success" >  <div style="text-align:justify;"> Propriété 2 :  <br>
    Il y ainsi 3 possibilités pour utiliser l'instruction <code> range </code> <br>
    
<code>for i in range(n): </code> le compteur i va de 0 à n-1 par pas de 1 : il y a donc bien n répétitions. <br>
<code>for i in range(m,n): </code> le compteur i va de m à n-1 par pas de 1 : il y a donc n-m répétitions. <br>
<code>for i in range(m,n,pas): </code> le compteur i va de m à n-1 par pas de pas ; le pas est un entier relatif non nul, il peut en particulier être négatif.    
    
</div>

<div class="alert alert-info"> <div style="text-align:justify;"> <font size="3"> <B> On souhaite tracer un carré de côté 100 pixels à l'aide d'une boucle for. On obtient le programme ci-dessous : </B></font>  
. <div> 

```python
from turtle import *
pencolor("red")
pensize(10)
goto(0,0)

for i in range(4):
    forward(100)
    left(90)
mainloop()


```

<div class="alert alert-info"> <div style="text-align:justify;"> <font size="3"> <B> Le programme ci-dessous permet de tracer un escalier de 4 marches. Modifier ce programme afin d'obtenir un escalier de 7 marches. </B></font>  


```python
from turtle import *



pencolor("red")
pensize(5)
penup()
goto(-200,-200)
pendown()
for i in range(4):
    left(90)
    forward(30)
    right(90)
    forward(30)
mainloop()
```

 ###  <p style="background-color:silver;"><b> <u>  b) Utilisation de la boucle while  </u> </b></p> 


<div class="alert alert-warning">   <div style="text-align:justify;"> La boucle répétitive conditionnelle while (tant que) est utilisée quand on ne connait pas le nombre de fois où l'on doit répéter une suite d'instructions. <br>
Dans la boucle while :
   <ul>
<li> la condition est évaluée une première fois </li>

<li> si la condition est Fausse, on n'entre pas dans la boucle et on continue après le bloc des instructions du while </li>

<li> si la condition est Vraie, on entre dans la boucle : les instructions du bloc des instructions du while sont exécutées </li>

<li> la condition est à nouveau évaluée et ainsi de suite, autant de fois que nécessaire jusqu'à ce que la condition soit fausse </li> </ul>

<div class="alert alert-success" >  <div style="text-align:justify;"> Propriété 1 :  <br>
La Syntaxe à utiliser avec l'instruction while :
    
<code>while condition : 
    instruction </code> 
</div>

<div class="alert alert-warning">   <div style="text-align:justify;"> Remarques :
   <ul>
<li> c'est l'indentation qui permet de sortir de la boucle </li>

<li> contrairment à la boucle for, l'initialisation et l'incrémentation du compteur ne sont pas gérée automatiquement par la boucle while </li>

<li> on utilise habituellement les lettres i, j pour les compteurs, mais ce n'est pas une obligation </li>

<li> on peut imbriquer plusieurs boucles while </li>
    </ul>

```python
from turtle import *
penup()
goto(-200,-200)
pendown()
i=0
while i<10:
    forward(20)
    left(90)
    forward(20)
    right(90)
    i=i+1


mainloop()
```

<div class="alert alert-warning">   <div style="text-align:justify;"> Remarques :
   <ul>
<li> à la ligne 5 on initialise le compteur à 0 </li>
<li> l'instruction while, à la ligne 6, indique à Python qu'il lui faut répéter continuellement le bloc d'instructions indenté qui suit (lignes 7 et 10), tant que la condition i < 10 est vraie </li>
       <li> ligne 11 on incrémente le compteur (ici on a un pas de 1) </li>
<li> lorsque la condition n’est plus vraie, le bloc indenté des lignes 7 à 10 est ignoré et le programme s'arrête </li>
    </ul>
